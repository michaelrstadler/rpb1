### 2. Concatenate stacks

```python
stack, frames = imp.concatenate_5dstacks([stack1, stack2, stack3, stack4, stack5, stack6, stack7])
viewer(stack, zmax=True)
```

### 3. Determine parameters for segmentation of nuclei.

```python
# Perform test initial segmentation of nuclei.
seed_window = (stack.shape[-3],40,40)
min_seed_dist = 25
size_min=0
size_max = 50000
circularity_min = 0
sigma=5

nucmask_test = imp.segment_nuclei_4dstack(stack[nucchannel,0:5], imp.segment_nuclei_3Dstack_rpb1, seed_window=seed_window, min_seed_dist=min_seed_dist, size_min=size_min, size_max=size_max, circularity_min=circularity_min, sigma=sigma)
```

#### Optional: view initial segmentation and ensure it is reasonable

```python
viewer(nucmask_test.astype(bool), figsize=12)
```

### 4. Perform initial segmentation and connect nuclei

```python
nucmask_init = imp.segment_nuclei_4dstack(stack[nucchannel], imp.segment_nuclei_3Dstack_rpb1, seed_window=seed_window, min_seed_dist=min_seed_dist, size_min=size_min, size_max=size_max, circularity_min=circularity_min, sigma=sigma)
nucmask_conn = imp.connect_nuclei(nucmask_init)
```

#### Optional: view connected segmentation

```python
viewer(nucmask_conn, figsize=12)
```

### 5. Interpolate to fill in gaps

```python
nucmask = imp.interpolate_nuclear_mask(nucmask_conn, 3)
#nucmask = nucmask_conn
```

#### View final nuclear segmentation

```python
viewer(nucmask, figsize=12)
```

### 6. Perform gaussian fitting on MS2 channel

```python
min_distances = (stack.shape[-2], 25, 25)
fitwindow_rad_xy = 10
fitwindow_rad_z = 2
sigma_small = 1

fits = imp.fit_ms2(stack[spotchannel], min_distances=min_distances, fitwindow_rad_xy=fitwindow_rad_xy, sigma_small=sigma_small)
```

### 7. Filter fits

```python
# Parameters for filtering fit
reload(imp)
peakiness = 4.2
fits_filtered = imp.filter_ms2fits(fits, peakiness, stack, spotchannel)
#imp.quickview_ms2(stack, fits_filtered, 1)
```

### 8. Connect spots from fits

```python
# Connect all frames.
spots_connected = imp.connect_ms2_frames_via_nuclei(fits_filtered, nucmask, nucmask_dilation=7)
```

```python
# Filter for trajectory length.
min_len = 10

spot_data = imp.filter_spot_duration(spots_connected, min_len)
```

```python
# Check the results.
imp.quickview_ms2(stack, spot_data, channel = spotchannel, MAX=True)
```

### Add corrected dataframes

```python
# Get z increment (in microns) from metadata.
metadata = czifile.CziFile(czi_file).metadata()
root = ET.fromstring(metadata)
z_increment = float(root[0][4][3][11][5][3][0][1].text)
print(z_increment)
```

```python
# Generate new dataframes using z depth correction calibration.
slope = -338
slice_thickness = z_increment
mv.intvol_zc, mv.intfit_zc, mv.prot_zc = imp.correct_spot_data_depth(mv.spot_data, slope=slope, slice_thickness=slice_thickness, 
    cols=[9,10,11], return_dfs=True)
```

```python
mv.intvol_bzc = imp.spotdf_bleach_correct(mv.intvol_zc, mv.stack[nucchannel])
mv.intfit_bzc = imp.spotdf_bleach_correct(mv.intfit_zc, mv.stack[nucchannel])
mv.prot_bzc = imp.spotdf_bleach_correct(mv.prot_zc, mv.stack[nucchannel])
```

```python
mv.intvol_bc = imp.spotdf_bleach_correct(mv.intvol, mv.stack[nucchannel])
mv.intfit_bc = imp.spotdf_bleach_correct(mv.intfit, mv.stack[nucchannel])
mv.prot_bc = imp.spotdf_bleach_correct(mv.prot, mv.stack[nucchannel])
```
